```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns=['class','v1','v2','v3','v4','v5','v6','v7','v8','v9','v10','v11','v12','v13']
df=pd.read_csv('winee.data', names=columns, delimiter=',')
```

```python
df.head()
```

```python
v1 = df["v1"].values
v2 = df["v2"].values
v3 = df["v3"].values
v4 = df["v4"].values
v5 = df["v5"].values
v6 = df["v6"].values
v7 = df["v7"].values
v8 = df["v8"].values
v9 = df["v9"].values
v10 = df["v10"].values
v11 = df["v11"].values
v12 = df["v12"].values
v13 = df["v13"].values
```

```python
X = np.array(list(zip(v1, v2, v3, v4)))
plt.scatter(v1, v2, v3, v4)
```

```python
def distance(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

```python
k = 5
C = X[0:5]
plt.scatter(v1, v2, v3, v4)
plt.scatter(C[:,0], C[:,1],C[:,2],C[:,3],marker="*")
```

```python
from copy import deepcopy

oldClusters = np.zeros(C.shape)

clusters = np.zeros(len(X))

error = dist(C, oldClusters, None)

count = 1
while (error !=0):
    for i in range(len(X)):
        distances = dist(X[i],C)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    
    oldClusters = deepcopy(C)
    
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j]==i]
        C[i] = np.mean(points, axis=0)
    error = dist(C, oldClusters, None)
```

```python
# colors = ['b', 'y', 'c', 'm','r']
fig, ax = plt.subplots()
for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=50, c=colors[i])
ax.scatter(C[:, 0], C[:, 1],C[:,2],C[:,3], marker='x')
```
